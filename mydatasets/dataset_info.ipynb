```python
import argparse
import os
import pickle
import numpy as np
```

```python
os.environ["CUDA_VISIBLE_DEVICES"] = "0"

parser = argparse.ArgumentParser()
parser.add_argument(
    "--dataset",
    default="retailRocket_DSAN",
    help=
    "dataset name: diginetica/retailRocket_DSAN/Tmall/Nowplaying",
)

opt = parser.parse_args(args=[])
```

```python
train_data_a = pickle.load(
        open("../datasets/" + opt.dataset + "/train.txt", "rb"))
test_data_a = pickle.load(
        open("../datasets/" + opt.dataset + "/test.txt", "rb"))
```

```python
def calculate_info(ds,len_sess_list,item_set):
    length_ds = len(ds)
    for i in range(length_ds):
        sess = ds[i]
        if not isinstance(ds[i],int):
            len_sess_list.append(len(sess))
            for x in sess:
                item_set.add(x)
        else:
            item_set.add(ds[i])
```

```python
len_sess_list_train = []
item_set_train = set()
item_set_train_y = set()
len_sess_list_test = []
item_set_test = set()
item_set_test_y = set()
calculate_info(train_data_a[0],len_sess_list_train,item_set_train)
calculate_info(train_data_a[1],[],item_set_train_y)
calculate_info(test_data_a[0],len_sess_list_test,item_set_test)
calculate_info(test_data_a[1],[],item_set_test_y)
all_sess_len = len_sess_list_train+len_sess_list_test
```

```python
avg_len_all = np.asarray(all_sess_len).mean()+1 # include label
# avg_len_train = np.asarray(len_sess_list_train).mean()
print(f'#train\t#test\t#item\#avg.len.')
num_items = len(item_set_train.union(item_set_test).union(item_set_train_y).union(item_set_test_y))
print(f'{len(train_data_a[0])}\t{len(test_data_a[0])}\t{num_items}\t{avg_len_all}')
```
